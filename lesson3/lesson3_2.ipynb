```python
import requests

# 將您的 curl 命令轉換成 requests
url = "http://localhost:11434/api/generate"

# 準備請求的資料 (JSON 格式)
data = {
    "model": "gemma3:1b",
    "prompt": "天空為什麼是藍的?",
    "stream": False
}

# 發送 POST 請求 (因為 curl 使用了 -d 參數)
response = requests.post(url, json=data)

# 檢查回應狀態
print(f"狀態碼: {response.status_code}")

# 取得回應內容
if response.status_code == 200:
    print("下載成功")
    
    # 只顯示 'response' 欄位的內容
    result = response.json()
    if 'response' in result:
        print("AI 回答：")
        
        # 使用 IPython.display 來渲染 Markdown
        from IPython.display import display, Markdown
        display(Markdown(result['response']))
        
        # 也可以同時顯示原始文字（可選）
        print("\n" + "="*50)
        print("原始文字版本：")
        print(result['response'])
    else:
        print("回應中沒有 'response' 欄位")
else:
    print(f"請求失敗: {response.text}")
```

```python
# 進階版本：純 Markdown 預覽模式

import requests
from IPython.display import display, Markdown

def get_ai_response(prompt, model="gemma3:1b"):
    """取得 AI 回應並以 Markdown 格式顯示"""
    
    url = "http://localhost:11434/api/generate"
    data = {
        "model": model,
        "prompt": prompt,
        "stream": False
    }
    
    try:
        response = requests.post(url, json=data)
        
        if response.status_code == 200:
            result = response.json()
            if 'response' in result:
                # 直接以 Markdown 格式顯示
                display(Markdown(f"## AI 回答\n\n{result['response']}"))
                return result['response']
            else:
                print("回應中沒有 'response' 欄位")
        else:
            print(f"請求失敗: {response.text}")
            
    except Exception as e:
        print(f"發生錯誤: {e}")
    
    return None

# 測試 Markdown 預覽
print("=== Markdown 預覽模式測試 ===")
ai_response = get_ai_response("請用 Markdown 格式寫一篇關於 Python 程式設計的簡介，包含標題、列表和程式碼範例")
```

```python
# 互動式 Markdown 預覽

def interactive_markdown_chat():
    """互動式 Markdown 聊天"""
    
    print("=== 互動式 Markdown 聊天 ===")
    print("輸入 'quit' 結束對話")
    
    while True:
        user_input = input("\n您：")
        
        if user_input.lower() == 'quit':
            print("再見！")
            break
            
        if user_input.strip():
            print("\nAI 正在思考...")
            
            url = "http://localhost:11434/api/generate"
            data = {
                "model": "gemma3:1b",
                "prompt": user_input,
                "stream": False
            }
            
            try:
                response = requests.post(url, json=data)
                
                if response.status_code == 200:
                    result = response.json()
                    if 'response' in result:
                        # 以 Markdown 格式顯示回應
                        display(Markdown(f"**AI：**\n\n{result['response']}"))
                    else:
                        print("AI 沒有回應")
                else:
                    print(f"請求失敗: {response.text}")
                    
            except Exception as e:
                print(f"發生錯誤: {e}")

# 取消註解來啟動互動式聊天
# interactive_markdown_chat()
```
